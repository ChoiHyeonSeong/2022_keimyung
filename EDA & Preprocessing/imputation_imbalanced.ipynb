```python
def imputing_1(data, input_cols, output_col):
  imputer = datawig.SimpleImputer(
    input_columns=input_cols, # impute에 사용할 컬럼을 지정해줄수 있다. 컬럼 '3'은 제외시켰다.
    output_column=output_col, # 컬럼 0의 missing value를 채운다.
    )
  imputer.fit(train_df=data, num_epochs=50)
  df_null_only = data[data[output_col].isnull()]
  np_imputed = imputer.predict(df_null_only)
  df_imputed = pd.DataFrame(np_imputed)
  data[output_col][data[output_col].isnull()] = df_imputed[output_col+'_imputed']
```

# 샘플링

```python
from sklearn.preprocessing import LabelEncoder
from imblearn.combine import SMOTETomek
```

```python
df = pd.read_csv('/content/drive/MyDrive/dataset/2022_겨울_학술_dataset/total_imputed.csv',index_col=0)
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
le = LabelEncoder()

df['NAME_CONTRACT_TYPE'] = le.fit_transform(df['NAME_CONTRACT_TYPE'])
df['CODE_GENDER'] = le.fit_transform(df['CODE_GENDER'])
```

```python
X_imb = df.drop(['TARGET','SK_ID_CURR'],axis=1)
y_imb = df['TARGET']
```

```python
smt = SMOTETomek(sampling_strategy='auto', random_state=42)
```

```python
X_samp, y_samp = smt.fit_resample(X_imb, y_imb)
```

```python
X_samp.to_csv('__1.csv')
y_samp.to_csv('__2.csv')
```
