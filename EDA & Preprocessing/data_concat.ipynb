```python
df_left_join.loc[index_, 'previous_AMT_ANNUITY'] = 0
df_left_join.loc[index_, 'previous_AMT_CREDIT'] = 0
df_left_join.loc[index_, 'previous_AMT_GOODS_PRICE'] = 0
df_left_join.loc[index_, 'previous_ID_COUNT'] = 0
df_left_join.loc[index_, 'previous_DAY_DECISION'] = 0
df_left_join.loc[index_, 'previous_REFUSED_COUNT'] = 0
df_left_join.loc[index_, 'previous_APPLICATION_TO_CREDIT'] = 0
```
