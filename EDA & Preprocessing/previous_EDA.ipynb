**previous_application_data 칼럼 설명**


---



[SK_ID_PREV] : 샘플의 대출과 관련된 홈 크레딧에 있는 이전 크레딧의 ID (샘플의 하나의 대출은 홈 크레딧에 0,1,2 이상의 이전 대출 신청이 있을 수 있다

[SK_ID_CURR] : 대출 ID

[NAME_CONTRACT_TYPE] :  계약상품유형 Cash loans, Consumer loans, Revolving loans, XNA

[AMT_ANNUITY] : 월 대출 지급액

[AMT_APPLICATION] : 대출 신청 금액

[AMT_CREDIT] : 대출금액(허가)

[AMT_DOWN_PAYMENT] : 대출 시 납부한 선금액

[AMT_GOODS_PRICE] : 소비자 대출상품액

[WEEKDAY_APPR_PROCESS_START] : 대출 신청 시작 요일

[HOUR_APPR_PROCESS_START] : 대출 신청 시작 시간대

[FLAG_LAST_APPL_PER_CONTRACT] : 이전 계약의 마지막 대출 신청 여부

[NFLAG_LAST_APPL_IN_DAY] : 하루중 마지막 대출 신청 여부(하루에 여러 번 대출 신청했을 경우)

[NFLAG_MICRO_CASH] : 소액 대출 여부

[RATE_DOWN_PAYMENT] : 선금 비율(정규화됨)

[RATE_INTEREST_PRIMARY] : 이자율

[RATE_INTEREST_PRIVILEGED] : 이자율

[NAME_CASH_LOAN_PURPOSE] : 현금 대출 목적

[NAME_CONTRACT_STATUS] : 대출 상태(허가, 취소)

[DAYS_DECISION] : 과거 신청 대비 현재 신청 결정 기간

[NAME_PAYMENT_TYPE] : 과거 대출 신청의 납부 방법

[CODE_REJECT_REASON] : 과거 신청 거절 사유

[NAME_TYPE_SUITE] : 동행 고객

[NAME_CLIENT_TYPE] : 신규 고객 또는 기존 대출 고객 여부

[NAME_GOODS_CATEGORY] : 대출 상품 중분류 유형

[NAME_PORTFOLIO] : 현금대출/POS/CAR 대출 유형

[NAME_PRODUCT_TYPE] : 고객이 찾아온 대출인가, 영업 대출인가[-

[CHANNEL_TYPE] : 채널 유형

[SELLERPLACE_AREA] : 판매자 판매 지역

[NAME_SELLER_INDUSTRY] : 판매자 Industry

[CNT_PAYMENT] : 이전 대출 신청의 대출금액 관련 Term

[NAME_YIELD_GROUP] : 집단 금리 적용 유형

[PRODUCT_COMBINATION] : 이전 대출 결합 상품

[DAYS_FIRST_DRAWING] : 신청날짜부터 최초 대출 지급까지의 일자

[DAYS_FIRST_DUE] : 신청날짜부터 마감일까지의 일자

[DAYS_LAST_DUE_1ST_VERSION] : 신청날짜부터 첫 만기일까지의 일자

[DAYS_LAST_DUE] : 신청날짜부터 마지막 만기일까지의 일자

[DAYS_TERMINATION] : 현 대출 신청일자 대비 대출 예상 종료 일자

[NFLAG_INSURED_ON_APPROVAL] : 대출 신청 시 보험가입 요청여부

## AMT_ANNUITY, AMT_CREDIT, AMT_GOODS_PRICE 평균값 추출

```python
previous_data = pd.DataFrame()
```

```python
mean_group = ['SK_ID_CURR', 'AMT_ANNUITY', 'AMT_CREDIT', 'AMT_GOODS_PRICE']
for i in mean_group:
  grouped_pv = pv_app_data.groupby('SK_ID_CURR')[i].mean()
  previous_data[('previous_' + i)] = pd.DataFrame(list(grouped_pv))
```

## 대출신청금액 대비 허가금액

```python
pv_app_data[pv_app_data['AMT_APPLICATION'] > pv_app_data['AMT_CREDIT']]
```

```python
id_list = []
value_list = []
for i in range(len(pv_app_data)):
  id_list.append(pv_app_data['SK_ID_CURR'][i])
  a = pv_app_data['AMT_APPLICATION'][i]
  b = pv_app_data['AMT_CREDIT'][i]
  c = b - a # 대출신청금액보다 대출허가금액이 더 클경우 1이상
  value_list.append(c)
APPLICATION_TO_CREDIT = pd.DataFrame(zip(id_list, value_list), columns=['SK_ID_CURR', 'APPLICATION_TO_CREDIT'])
```
